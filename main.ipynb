# 5 - KeyPhrases Mapping

```python
sentences = KeyPhrases_Mapping.tokenize_sentences(summerized_article)
```

```python
keyword_sentence_mapping = KeyPhrases_Mapping.get_sentences_for_keyword(filters, sentences)
```

```python
for key in keyword_sentence_mapping:
    try:
        print(f"Keyword ({key}) it's paragraph:\n{keyword_sentence_mapping[key][0]}")
        print("="*100)
        print("\n")
    except:
        pass
```

<a id='6'></a>
# 6 - Create a DF for Each Phrase and it's Paragraph

```python
paragraphs = []
phrases = []

for key in keyword_sentence_mapping:
    try:
        paragraphs.append(keyword_sentence_mapping[key][0])
        phrases.append(key)
    except:
        pass
```

```python
for i in paragraphs:
    print(i)
    print()
```

```python
key_dics = {
    'Phrases':phrases,
    'Paragraphs':paragraphs
}

key_dics
```

```python
keyPhrase_DF = pd.DataFrame(data=key_dics)
```

```python
keyPhrase_DF
```

```python
keyPhrase_DF.to_csv("KeyPhrases_DataFrame.csv")
```
