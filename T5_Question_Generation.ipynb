#**T5 Models**

```python
def T5_singleTask_questioGeneration(paragraphs):
  answers = []
  questions = []
  nlp = pipeline("question-generation", model="valhalla/t5-base-qg-hl")
  try:
      for x in paragraphs:
        qg = nlp(x)
        print(f"Question: {qg[0]['question']}")
        print(f"Answer: {qg[0]['answer']}")
        print("="*100)
        answers.append(qg[0]['answer'])
        questions.append(qg[0]['question'])
  except:
      pass

  return answers, questions


def T5_multiTaskSmallModel_questioGeneration(paragraphs):
  answers = []
  questions = []
  nlp = pipeline("multitask-qa-qg")
  try:
      for x in paragraphs:
        qg = nlp(x)
        print(f"Question: {qg[0]['question']}")
        print(f"Answer: {qg[0]['answer']}")
        print("="*100)
        answers.append(qg[0]['answer'])
        questions.append(qg[0]['question'])
  except:
      pass

  return answers, questions

def T5_multiTaskBaseModel_questioGeneration(paragraphs):
  answers = []
  questions = []
  nlp = pipeline("multitask-qa-qg", model="valhalla/t5-base-qa-qg-hl")
  try:
      for x in paragraphs:
        qg = nlp(x)
        print(f"Question: {qg[0]['question']}")
        print(f"Answer: {qg[0]['answer']}")
        print("="*100)
        answers.append(qg[0]['answer'])
        questions.append(qg[0]['question'])
  except:
      pass

  return answers, questions



def T5_e2eSmallModel_questioGeneration(paragraphs):
  answers = []
  questions = []
  nlp = pipeline("e2e-qg")
  try:
      for x in paragraphs:
        qg = nlp(x)
        print(f"Question: {qg[0]['question']}")
        print(f"Answer: {qg[0]['answer']}")
        print("="*100)
        answers.append(qg[0]['answer'])
        questions.append(qg[0]['question'])
  except:
      pass

  return answers, questions


def T5_e2eBaseModel_questioGeneration(paragraphs):
  answers = []
  questions = []
  nlp = pipeline("e2e-qg", model="valhalla/t5-base-e2e-qg")
  try:
      for x in paragraphs:
        qg = nlp(x)
        print(f"Question: {qg[0]['question']}")
        print(f"Answer: {qg[0]['answer']}")
        print("="*100)
        answers.append(qg[0]['answer'])
        questions.append(qg[0]['question'])
  except:
      pass

  return answers, questions
```

```python
def createDataFrame(questions, answers):
  QG_dataFrame = {
    "Questions" : questions,
    "Answers" : answers
  }
  QG_dataFrame = pd.DataFrame(QG_dataFrame)
  return QG_dataFrame
```

###***3.1 - Single Task QG***

```python
t5_singleTask_answers , t5_singleTask_questions = T5_singleTask_questioGeneration(paragraphs)
```

```python
t5_singleTaskQG_DataFrame = createDataFrame(t5_singleTask_questions, t5_singleTask_answers)
```
